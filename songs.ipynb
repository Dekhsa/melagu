```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import os

# --- KONFIGURASI ---
# ⚠️ PASTIKAN ANDA MENGISI INI DENGAN BENAR DARI RAPIDAPI
RAPIDAPI_KEY = "ad902bdf3fmshe0c651151169560p1b1074jsn07e9ade21b53"
RAPIDAPI_HOST = "genius-song-lyrics1.p.rapidapi.com"

INPUT_CSV = "songs.csv"
OUTPUT_CSV = "songs_with_lyrics_genius_final.csv"

# --- Muat Data ---
try:
    df = pd.read_csv(INPUT_CSV)
    if 'lyrics' not in df.columns:
        df['lyrics'] = None
    print(f"Memulai dari file '{INPUT_CSV}'.")
except FileNotFoundError:
    print(f"Error: File '{INPUT_CSV}' tidak ditemukan.")
    exit()

# --- PROSES API + SCRAPING ---
api_headers = {
    "X-RapidAPI-Key": RAPIDAPI_KEY,
    "X-RapidAPI-Host": RAPIDAPI_HOST
}

# Header untuk scraping, menyamar sebagai browser
scraping_headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

for index, row in df.iterrows():
    if pd.notna(row['lyrics']):
        continue

    try:
        # 1. GUNAKAN API UNTUK MENCARI URL
        search_url = f"https://{RAPIDAPI_HOST}/search/"
        query = f"{row['song_name']} {row['song_artist']}"
        querystring = {"q": query, "per_page": "1"}
        
        api_response = requests.get(search_url, headers=api_headers, params=querystring)
        api_response.raise_for_status()
        api_data = api_response.json()
        
        if not api_data.get('hits'):
            df.at[index, 'lyrics'] = "NOT_FOUND_IN_API"
            print(f"✗ Tidak ditemukan di API: {query}")
            continue

        song_url = api_data['hits'][0]['result']['url']
        print(f"✓ URL ditemukan: {song_url}")

        # 2. GUNAKAN SCRAPER DENGAN HEADER BARU
        # ================== PERUBAHAN DI SINI ==================
        page_response = requests.get(song_url, headers=scraping_headers)
        # =======================================================
        page_response.raise_for_status()
        
        soup = BeautifulSoup(page_response.text, 'html.parser')
        lyrics_div = soup.find("div", attrs={"data-lyrics-container": "true"})
        
        if lyrics_div:
            lyrics = lyrics_div.get_text(separator='\n').strip()
            df.at[index, 'lyrics'] = lyrics
            print(f"✓ Lirik berhasil di-scrape untuk: {query}")
        else:
            df.at[index, 'lyrics'] = "LYRICS_DIV_NOT_FOUND"
            print(f"✗ Div lirik tidak ditemukan di halaman: {query}")

    except requests.exceptions.HTTPError as http_err:
        print(f"✗ HTTP Error pada: {query}. Detail: {http_err}")
        df.at[index, 'lyrics'] = f"HTTP_ERROR_{http_err.response.status_code}"
    except Exception as e:
        df.at[index, 'lyrics'] = "ERROR"
        print(f"✗ Error lain pada: {query}. Detail: {e}")

    # Simpan kemajuan
    if (index + 1) % 10 == 0:
        df.to_csv(OUTPUT_CSV, index=False)
        print(f"--- Kemajuan disimpan (baris ke-{index+1}) ---")

    time.sleep(1.5)

# Simpan hasil akhir
df.to_csv(OUTPUT_CSV, index=False)
print(f"🚀 Selesai! Hasil disimpan di '{OUTPUT_CSV}'.")
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import os
import re

# --- KONFIGURASI ---
RAPIDAPI_KEY = "ad902bdf3fmshe0c651151169560p1b1074jsn07e9ade21b53"
RAPIDAPI_HOST = "genius-song-lyrics1.p.rapidapi.com"

INPUT_CSV = "songs.csv"
OUTPUT_CSV = "songs_with_lyrics.csv"

# --- Muat Data atau Lanjutkan Progres ---
try:
    if os.path.exists(OUTPUT_CSV):
        df = pd.read_csv(OUTPUT_CSV)
        print(f"Melanjutkan dari file '{OUTPUT_CSV}'. Progres awal: {df['lyrics'].notna().sum()} dari {len(df)} lagu.")
    else:
        df = pd.read_csv(INPUT_CSV)
        if 'lyrics' not in df.columns:
            df['lyrics'] = None
        print(f"Memulai dari awal dengan file '{INPUT_CSV}'.")
except FileNotFoundError:
    print(f"Error: File '{INPUT_CSV}' tidak ditemukan.")
    exit()

# Dapatkan total lagu untuk info progres
total_songs = len(df)

# --- PROSES UTAMA ---
api_headers = {"X-RapidAPI-Key": RAPIDAPI_KEY, "X-RapidAPI-Host": RAPIDAPI_HOST}
scraping_headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

for index, row in df.iterrows():
    # Lewati baris yang sudah ada isinya (selain "unknown" atau error markers lama)
    if pd.notna(row['lyrics']) and row['lyrics'] != 'unknown':
        continue

    query = f"{row['song_name']} {row['song_artist']}"
    # Info Progres
    print(f"({index + 1} / {total_songs}) Mencari: {query}")

    try:
        # 1. Panggil API untuk mencari URL
        search_url = f"https://{RAPIDAPI_HOST}/search/"
        querystring = {"q": query, "per_page": "1"}
        
        api_response = requests.get(search_url, headers=api_headers, params=querystring)
        
        if api_response.status_code == 429:
            print("--- Rate Limit Tercapai. Jeda 60 detik... ---")
            time.sleep(60)
            api_response = requests.get(search_url, headers=api_headers, params=querystring)
            
        api_response.raise_for_status()
        api_data = api_response.json()
        
        if not api_data.get('hits'):
            df.at[index, 'lyrics'] = "unknown"
            print(f"✗ Tidak ditemukan di API: {query}")
            continue

        song_url = api_data['hits'][0]['result']['url']
        
        # 2. Scrape URL
        page_response = requests.get(song_url, headers=scraping_headers)
        page_response.raise_for_status()
        
        soup = BeautifulSoup(page_response.text, 'html.parser')
        lyrics_div = soup.find("div", attrs={"data-lyrics-container": "true"})
        
        if lyrics_div:
            raw_lyrics = lyrics_div.get_text(separator='\n').strip()
            if "Lyrics" in raw_lyrics:
                raw_lyrics = raw_lyrics.split("Lyrics", 1)[-1]
            
            cleaned_lyrics = re.sub(r'\[.*?\]', '', raw_lyrics)
            cleaned_lyrics = os.linesep.join([s for s in cleaned_lyrics.splitlines() if s])
            
            df.at[index, 'lyrics'] = cleaned_lyrics.strip()
            print(f"✓ Lirik berhasil di-scrape untuk: {query}")
        else:
            df.at[index, 'lyrics'] = "unknown"
            print(f"✗ Div lirik tidak ditemukan di halaman: {query}")

    except Exception as e:
        print(f"✗ Terjadi Error pada: {query}. Detail: {e}")
        # ================== PENYIMPANAN OTOMATIS SAAT ERROR ==================
        print("--- Terjadi Error. Menyimpan progres terakhir... ---")
        df.to_csv(OUTPUT_CSV, index=False)
        # =====================================================================
        df.at[index, 'lyrics'] = "unknown" # Tandai sebagai unknown setelah error
        # Jika error karena API key, hentikan proses
        if '403' in str(e) or '401' in str(e):
             print("Proses dihentikan karena kesalahan Kunci API.")
             break

    # Jeda antar permintaan
    time.sleep(2)

# Simpan hasil akhir setelah loop selesai normal
df.to_csv(OUTPUT_CSV, index=False)
print(f"🚀 Selesai! Hasil akhir disimpan di '{OUTPUT_CSV}'.")
```
